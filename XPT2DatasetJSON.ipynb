```python
import requests
import xmltodict
import json
import pandas as pd
```

```python
def xml_to_json(xml_url):
  try:
    response = requests.get(xml_url)
    response.raise_for_status()  # Raise an exception for bad status codes

    xml_content = response.content
    json_data = xmltodict.parse(xml_content)
    return json_data

  except requests.exceptions.RequestException as e:
    print(f"Error fetching XML: {e}")
    return None
  except xmltodict.ParsingInterrupted as e:
    print(f"Error parsing XML: {e}")
    return None

```

```python
metagdict = {}
metadict = {}

obj = xml_to_json("https://raw.githubusercontent.com/cdisc-org/sdtm-adam-pilot-project/refs/heads/master/updated-pilot-submission-package/900172/m5/datasets/cdiscpilot01/tabulations/sdtm/define.xml")

for x in obj["ODM"]["Study"]["MetaDataVersion"]["ItemGroupDef"]:
  OID      = x['@OID']
  LABEL    = x['@def:Label']
  metagdict[OID] = LABEL

for x in obj["ODM"]["Study"]["MetaDataVersion"]["ItemDef"]:
  OID      = x['@OID']
  NAME     = x['@Name']
  LABEL    = x['@def:Label']
  DATATYPE = x['@DataType']
  LENGTH   = x['@Length']
  metadict[OID] = [OID, NAME, LABEL ,DATATYPE, LENGTH]

metadict['AE.STUDYID']
```

```python
# prompt: 現在時刻をISO8601形式で取得する

from datetime import datetime, timezone

def get_current_datetime_iso8601():
  """Gets the current date and time in ISO 8601 format."""
  now = datetime.now(timezone.utc)  # Use UTC for consistency
  return now.isoformat()

current_datetime_iso8601 = get_current_datetime_iso8601()
print(current_datetime_iso8601)

SDTM_JSON = {}
SDTM_JSON["datasetJSONCreationDateTime"] = current_datetime_iso8601
```

```python
SDTM_JSON ={}


def xpt2datasetjson(xptName):
  url = "https://github.com/cdisc-org/sdtm-adam-pilot-project/raw/refs/heads/master/updated-pilot-submission-package/900172/m5/datasets/cdiscpilot01/tabulations/sdtm/" + xptName + ".xpt"
  df = pd.read_sas(url, encoding="ISO-8859-1")

  SDTM_JSON["datasetJSONCreationDateTime"] = get_current_datetime_iso8601()
  SDTM_JSON["datasetJSONVersion"] = "1.1.0"
  SDTM_JSON["fileOID"] = "JPMA/2024TF1-2/DataPreparation"
  SDTM_JSON["dbLastModifiedDateTime"] = "2024-12-12T00:00:00"
  SDTM_JSON["originator"] = "CDISC"
  SDTM_JSON["sourceSystem"] = {"name": "Google Colaboratory", "version": "2024-12-12T00:00:00"}
  SDTM_JSON["studyOID"] = "cdisc.com/CDISCPILOT01"
  SDTM_JSON["metaDataVersionOID"] = "CDISC.SDTMIG.3.1.2"
  SDTM_JSON["metaDataRef"] = "https://github.com/cdisc-org/sdtm-adam-pilot-project/blob/master/updated-pilot-submission-package/900172/m5/datasets/cdiscpilot01/tabulations/sdtm/define.xml"
  SDTM_JSON["itemGroupOID"] = "IG." + xptName.upper()
  SDTM_JSON["records"] = len(df)
  SDTM_JSON["name"] = xptName.upper()
  SDTM_JSON["label"] = metagdict[xptName.upper()]

  items = []
  for column in df.columns:
    itemmeta = {
        "itemOID": "IT." + metadict[xptName.upper()+"."+column][0],
        "name": metadict[xptName.upper()+"."+column][1],
        "label":metadict[xptName.upper()+"."+column][2],
        "dataType": metadict[xptName.upper()+"."+column][3],
        "length": metadict[xptName.upper()+"."+column][4]
    }
    items.append(itemmeta)

  SDTM_JSON["columns"] = items
  SDTM_JSON["rows"] = df.values.tolist()

  with open(xptName + ".json", "w") as f:
    json.dump(SDTM_JSON, f)  # Use indent for pretty printing

  return

```

```python
for domain in metagdict.keys():
  xpt2datasetjson(domain.lower())
```
